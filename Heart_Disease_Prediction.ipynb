Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Data Collection and Processing

```python
#loading a csv data into Pandas dataframe
heart_data=pd.read_csv(r'C:\Users\Udaya\Downloads\heart.csv')
```

```python
#getting info about the dataset
heart_data.info()
```

```python
#checking for missing values
heart_data.isnull().sum()
```

```python
#statistical measures about the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data['target'].value_counts()
```

1 represents diseased heart.
0 represents healthy heart.

```python
#Splitting the features and target(label)
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training - Using Logistic Regression

```python
model=LogisticRegression()
```

```python
#training the model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data : ",training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data : ",test_data_accuracy)
```

Building a Predictive System

```python
input_data=(55,1,0,132,353,0,1,132,1,1.2,1,1,3)

#changing the input data to a numpy array
input_data_as_numoy_array=np.asarray(input_data)

#reshape the numpy array as we are predicting for only 1 instance
input_data_reshaped = input_data_as_numoy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if(prediction[0] == 0):
    result_text="The person does NOT have any heart disease"
    bar_color='green'
    bar_values=[1,0]
else:
    result_text="The person HAS heart disease"
    bar_color='red'
    bar_values=[0,1]

print(result_text)

```

```python
#Plotting a bar graph
feature_names = ["Healthy Heart", "Diseased Heart"]

plt.figure(figsize=(5,3))
plt.bar(feature_names, bar_values, color=bar_color, width=0.3)
plt.title("Heart Disease Prediction\n\n" + result_text, fontsize=12)
plt.yticks([])
plt.ylim(0, 1.2)
plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.tight_layout()
plt.show()
```
